```python
import numpy as np
from sklearn.cluster import MeanShift
```

```python
df_exercises[df_exercises['workload'].isin([np.nan])]
```

```python
df_exercises = df_exercises.dropna(how='any')
```

```python
df_exercises
```

```python
df_exercises = df_exercises[~(df_exercises == 0).any(axis=1)]
df_exercises
```

```python
ms = MeanShift()
ms.fit(df_exercises.drop(['splitName' ],axis = 1))
labels = ms.labels_
cluster_centers = ms.cluster_centers_

print(cluster_centers)

n_clusters_ = len(np.unique(labels))

print("Number of estimated clusters:", n_clusters_)

print(labels)
```

```python
len(labels)
```

```python
df_exercises = df_exercises.reset_index(drop=True)
df_exercises['labels'] = labels
df_exercises
```

```python
drills = df_exercises['splitName'].unique()
```

```python
columns = ['splitName']
for cluster in range(n_clusters_):
    columns.append('cat' + str(cluster + 1))
```

```python
columns
```

```python
drill_clusters = pd.DataFrame()

for drill in drills:
    df_specific_drill = df_exercises.loc[df_exercises['splitName'].isin([drill])]
    
    df_value_counts = pd.DataFrame(df_specific_drill['labels'].value_counts(normalize=True).sort_index())
    
    new_row = []
    
    for i in range(len(columns[1:])):
        try:
            new_row.append(df_value_counts.loc[i]['labels'])
        except:
            new_row.append(0)
            
    new_row.insert(0, drill)
                    
    new_row = pd.Series(new_row)

    drill_clusters = drill_clusters.append(new_row, ignore_index=True)
```

```python
len(drills)
```

```python
drill_clusters.columns = columns
```
